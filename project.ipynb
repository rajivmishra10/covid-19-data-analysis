```python
import pandas as pd
import numpy as np
import folium
from geopy.geocoders import Nominatim
```

```python
## I am takinf covid 19 data set
```

```python
df_covid=pd.read_csv('owid-covid-data.csv')
```

```python
df_covid.shape
```

```python
df_covid.head()
```

```python
## Im taking specific data of 2020-05-24 and rest im deleting from my data frame
```

```python
df_covid=df_covid.drop(df_covid[df_covid['date'] != "2020-05-24"].index)
```

```python
df_covid=df_covid.drop(df_covid[df_covid['location'] == "World"].index)
```

```python
##Im making a copy of my data set for making choropleth map
## And in data frame im taking top 12 countries with most covid cases
```

```python
df_covid_c=df_covid
df_covid=df_covid.nlargest(12, ['total_cases'])
```

```python
df_covid
```

```python
df_covid=df_covid.fillna(0)
```

```python
df_covid
```

```python
df_covid.describe()
```

```python
df_covid
```

```python
## Here im calculating a round no of death percentage
```

```python
df_covid_c["death_percentage"]=round((df_covid_c["total_deaths"]/df_covid_c["total_cases"])*100)
```

```python
df_covid
```

```python
ar=df_covid["iso_code"].to_numpy()
```

```python
ar
```

```python
## here im collecting longtidu and latitude of countries from api to create bubble map
```

```python
import time
ar_lo = []
ar_la = []
geolocator = Nominatim()
for x in ar:
    time.sleep( 2 )
    try:
        loc = geolocator.geocode(x[:2])
        print(loc)
        ar_lo.append(loc.longitude)
        ar_la.append(loc.latitude)
    except:
        ar_lo.append(None)
        ar_la.append(None)
        
    
```

```python

s_long = pd.Series(ar_lo,name='long')
s_late = pd.Series(ar_la,name="late")
```

```python
s_long
ar_lo
```

```python
df_covid.reset_index(inplace=True)
```

```python
df_covid=pd.concat([df_covid,s_late,s_long], axis=1)
```

```python
df_covid
```

```python
m = folium.Map(location=[0,0], tiles="Stamen Toner", zoom_start=2)
```

```python
## Here im creating a bubble map with most cases country wise
```

```python
for i in range(0,len(df_covid)):
    if np.isnan(df_covid.iloc[i]['long']):
        continue
    folium.Circle(
      geo_data="custom.geo.json",
      location=[ar_la[i],ar_lo[i]],
      popup="Deaths are "+str(df_covid.iloc[i]['total_deaths']),
      radius=int(df_covid.iloc[i]['total_cases']),
      color='crimson',
      fill=True,
      fill_color='crimson'
   ).add_to(m)
```

```python
m.save('mymap.html')
```

```python
m2 = folium.Map(location=[0,0], zoom_start=1.5)
```

```python
## Here im creating choropleth map of all countries with the data of death percentage 
```

```python
folium.Choropleth(
    geo_data="custom.geo.json",
    name='choropleth',
    data=df_covid_c,
    columns=['location','death_percentage'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.9,
    line_opacity=0.5,
    bins=[0, 4, 8, 12, 16, 20, 24, 28],
    legend_name= 'death rate percentage'
).add_to(m2)

folium.LayerControl().add_to(m2)
```

```python
m2.save('mymap2.html')
```
